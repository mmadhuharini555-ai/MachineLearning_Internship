```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df = pd.read_csv("/content/drive/MyDrive/INTERNSHIP/Madhu/Apriori/Groceries_dataset.csv")
df.isnull().sum()
```

```python
transactions = (
    df.groupby(["Member_number", "Date"])["itemDescription"]
      .apply(list)
      .tolist()
)
```

```python
all_items = sorted(set(item for t in transactions for item in t))
data = []
for t in transactions:
    row = {item: 0 for item in all_items}
    for item in t:
        row[item] = 1
    data.append(row)

basket = pd.DataFrame(data)
```

```python
MIN_SUPPORT = 0.02      # appears in at least 2% of transactions
MIN_CONFIDENCE = 0.3    # rules must have >= 30% confidence
```

```python
frequent_itemsets = fpgrowth(
    basket,
    min_support=MIN_SUPPORT,
    use_colnames=True
)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=MIN_CONFIDENCE
).sort_values("lift", ascending=False)
```

```python
print("Top 10 rules:")
print(rules[["antecedents", "consequents", "support", "confidence", "lift"]].head(10))
```

```python
# --- Plot: support vs confidence (color = lift) ---
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    rules["support"],
    rules["confidence"],
    c=rules["lift"],
    cmap="viridis",
    alpha=0.7
)
plt.colorbar(scatter, label="Lift")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Association Rules (FPGrowth)")
plt.show()
```
