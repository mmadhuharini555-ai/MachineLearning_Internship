```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
ta=pd.read_csv('/content/drive/MyDrive/INTERNSHIP/Madhu/K_Means_Clustering/futuristic_city_traffic.csv')
```

```python
ta.head()
```

```python
ta.isnull().sum()
```

```python
x=ta[['Is Peak Hour','Random Event Occurred']]
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(x)
labels=kmeans.labels_
print(labels)
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=labels)
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    marker='X',
    s=200
)
plt.xlabel("Peak Hour")
plt.ylabel("Random Incident Occurred")
plt.title("K-Means Clustering_TrafficAnalysis")
plt.show()
```
