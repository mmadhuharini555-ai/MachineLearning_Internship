```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
# Create augmentation generator
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Load single image
img = load_img('/content/drive/MyDrive/INTERNSHIP/Madhu/Data_Augmentation/WhatsApp Image 2025-12-10 at 3.24.29 PM.jpeg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)
```

```python
# Generate 20 augmented images
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir="/content/drive/MyDrive/INTERNSHIP/Madhu/Data_Augmentation",
                          save_prefix="aug", save_format="jpg"):
    i += 1
    if i > 20:
        break
```
