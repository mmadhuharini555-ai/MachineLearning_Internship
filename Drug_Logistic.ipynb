```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dg=pd.read_csv('/content/drive/MyDrive/INTERNSHIP/Madhu/Linear&Logistic/Drug_logistic.csv')
dg.head()
```

```python
dg.isnull().sum()
```

```python
ind=dg[['Na_to_K']]
dep=dg[['Drug']]
```

```python
logr=LogisticRegression()
logr.fit(ind,dep)
```

```python
Na_to_K=float(input("Enter Na_to_K level:"))
pred=logr.predict([[Na_to_K]])
print(pred)
```

```python
logr.score(ind,dep)
```

```python
from sklearn.metrics import accuracy_score
pval=logr.predict(ind)
accuracy_score(dep,pval)
```

```python
import matplotlib.pyplot as plt
plt.scatter(ind, dep['Drug'])
plt.xlabel("Na_to_K")
plt.ylabel("Drug")
plt.title("Drug classification")
plt.show()
```
