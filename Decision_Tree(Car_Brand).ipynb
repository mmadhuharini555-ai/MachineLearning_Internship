```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
cards=pd.read_csv('/content/drive/MyDrive/INTERNSHIP/Madhu/Decision_Tree/cars.csv')
```

```python
le=LabelEncoder()
```

```python
y = le.fit_transform(cards[' brand'])
```

```python
x=cards[[' cylinders']]
y=le.transform(cards[' brand'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
cy=int(input("Enter cylinder(4 or 8):"))
pred=model.predict([[cy]])
print(pred)
```

```python
from sklearn.tree import plot_tree
plot_tree(model,filled=True,max_depth=5)
```
