```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
hp=pd.read_excel('/content/drive/MyDrive/INTERNSHIP/Madhu/Linear&Logistic/Houseprice_linear.xlsx')
hp.head()
```

```python
hp.isnull().sum()
```

```python
x=hp[['Area_sqft','Bedrooms','Age','Location_Score']]
y=hp['Price_Lakhs']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
print(pred)
```

```python
print("Intercept:",LR.intercept_)
print("Coeffients:",LR.coef_)
```

```python
print("MSE:",mean_squared_error(y_test,pred))
print("R2:",r2_score(y_test,pred))
```

```python
new_house=[[1000,2,4,8]]
pred_price=LR.predict(new_house)
print("Predicted Price:",pred_price)
```

```python
plt.scatter(y_test, pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black', linestyle='--') # Line for perfect prediction
plt.xlabel("Actual Price (Lakhs)")
plt.ylabel("Predicted Price (Lakhs)")
plt.title("House Price Prediction")
plt.show()
```
