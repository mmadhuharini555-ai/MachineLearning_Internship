```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
```

```python
sa=pd.read_csv('/content/drive/MyDrive/INTERNSHIP/Madhu/K_Nearest_Neighbor/enhanced_anxiety_dataset.csv')
```

```python
sa.head()
```

```python
sa.isnull().sum()
```

```python
x=sa[['Stress Level (1-10)','Heart Rate (bpm)','Breathing Rate (breaths/min)','Sweating Level (1-5)']]
y=sa['Anxiety Level (1-10)']
```

```python
k=5
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(x,y)
```

```python
stl=int(input("Enter Stress level(1-10):"))
hr=int(input("Enter Heart Rate(bpm):"))
br=int(input("Enter Breathing Rate(breaths/min):"))
swl=int(input("Enter Sweat level(1-5):"))
prediction=knn.predict([[stl,hr,br,swl]])
print("Predicted Anxiety Level is:",prediction)
```

```python
plt.plot(x,y)
plt.xlabel("Stress & Sweat Level,Heart & Breathing Rate")
plt.ylabel("Anxiety Level")
plt.title("K-Neighor_SocialAnxiety")
plt.show()
```
