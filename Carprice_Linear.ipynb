```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
cp=pd.read_csv('/content/drive/MyDrive/INTERNSHIP/Madhu/Linear&Logistic/Carprice_linear.csv')
```

```python
cp.head()
```

```python
cp.info()
```

```python
cp.isnull().sum()
```

```python
inp=[['Engine Size']]
out=['Price']
```

```python
LR=LinearRegression()
```

```python
LR.fit(cp[inp[0]],cp[out])
```

```python
LR.predict([[5.5]])
```

```python
plt.scatter(cp[inp[0]], cp[out])
plt.plot(cp[inp[0]], LR.predict(cp[inp[0]]), color='black')
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.title("Car Price Prediction")
plt.show()
```
