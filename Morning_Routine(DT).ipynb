```python
mr['Sleep Duration (mins)']=(mr['Sleep Duration (hrs)']*60).astype(int)
print(mr[['Sleep Duration (hrs)','Sleep Duration (mins)']].head())
```

```python
x=mr[['Sleep Duration (mins)','Meditation (mins)','Exercise (mins)']]
y=mr['Productivity Score (1-10)']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
sp=float(input("Enter Sleep Duration(hrs):"))
med=int(input("Enter Meditation Duration(mins):"))
ex=int(input("Enter Exercise Duration(mins):"))
pred=model.predict([[sp,med,ex]])
print("Productivity Score:",pred)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
model = DecisionTreeClassifier(
    max_depth=2,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
model.fit(x_train, y_train)
plt.figure(figsize=(20, 10))
plot_tree(
    model,
    filled=True,
    max_depth=2,
    fontsize=10
)
plt.show()
```
